```python

import matplotlib.pyplot as plt
import numpy as np
from autograd import grad, jacobian, hessian
import math
from numpy.linalg import norm
```

```python
def f(x, y):
  return (1-np.cos(0.5 * np.pi * x))*(y**2)

def gradient(x, y):
  x_gradient = (y**2/2) * np.pi  * np.sin((np.pi/2 * x))
  y_gradient = 2*y*(1-np.cos(np.pi * x /2))
  #mag = np.sqrt(x_gradient**2 + y_gradient**2)
  return np.array([x_gradient, y_gradient])
  
x = np.linspace(-1, 1, 50)
y = np.linspace(-1, 1, 50)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
plt.contourf(X, Y, Z,70, cmap='RdGy')
plt.colorbar()
```

```python

def gd_trajectory(num_initial):
    plt.axis('equal')
    plt.contourf(x,y,Z,200,cmap=plt.cm.RdBu)     
    maxIter = 10
    for i in range(num_initial):
        r_gd_iterates = np.zeros((2,maxIter))
        f_value = np.zeros(maxIter)
        r = np.random.uniform(-1,1,2) # initialization
        for iter in range(maxIter):
            r = r -  gradient(r[0], r[1])        
            r_gd_iterates[:,iter] = r      
            f_value[iter] = f(r[0], r[1])

     
        plt.plot(r_gd_iterates[0,:],r_gd_iterates[1,:],c="black", marker = '.', linewidth = 0.8)




    


    
```

```python
gd_trajectory(5)
```

```python
gd_trajectory(1000)
```
